```python
type(churn_data)
```

```python
plt.pie(churn_data, labels= ['Not Chunred', 'Churned'], colors=['Orange','Lime'], startangle=50, shadow=True, radius=1, explode=(0,0.2), autopct="%1.2f%%", pctdistance=0.75)
circle = plt.Circle((0,0), 1, color="Red")
c = plt.gcf()
c.gca().add_artist(circle)
plt.title("CHurned vs Not Churned")
plt.show()
```
